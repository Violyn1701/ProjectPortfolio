```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# IMPORT DATASET
path = r"C:\Users\violy\OneDrive\Desktop\Data_Analyst\Alex_The_Analyst\Movie_Industry\movies.csv"
raw_data = pd.read_csv(path)
df = raw_data.copy()
df
```

```python
# DATASET SUMMARY
def summary(df):
    print("\nNumber of rows                              :\t", df.shape[0])
    print("\nNumber of columns                         :\t", df.shape[1])
    print("\nNumber of duplicated rows              :\t", df.duplicated().sum())
    print("\nDataSet Information                         :")
    display(df.info())
    print("\nDataSet Description                         :")
    display(df.describe())

summary(df)
```

```python
# SUMMARY UNIQUENESS
def unique_sum(df):
    for i in df.columns:
        print(i)
        display(df[i].unique())
        print("")

unique_sum(df)
```

```python
# RELEASE COLUMN
df = df.dropna(subset = ["released"])
df['released'] = df['released'].str.replace(r'\s*\([^)]*\)', '', regex=True)
df["released"] = pd.to_datetime(df["released"], format="mixed")
df["released"]
```

```python
df = df.dropna(subset = ["score"])
summary(df)
```

```python
# REST OF MISSING VALUES
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
df['runtime'] = df.groupby('genre')['runtime'].transform(
    lambda x: x.fillna(x.median()))
df["writer"] = df["writer"].fillna("Unknown")
df["star"] = df["star"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["company"] = df["company"].fillna("Unknown")
summary(df)
```

```python
# FILLING IN MISSING VALUES WITH MEDIAN, LESS AFFECTED BY OUTLIERS
df["budget"] = df["budget"].fillna(df["budget"].median())
df["gross"] = df["gross"].fillna(df["gross"].median())
summary(df)
```

```python
# DATA TYPES
df["gross"] = df["gross"].astype("int64")
df["budget"] = df["budget"].astype("int64")
summary(df)
```

```python
# CORRECT RELEASE YEAR
df["year"] = df["released"].dt.year
df
```

```python
# RESETTING INDEX
df = df.reset_index(drop = True)
df
```

```python
# Sort by gross
df.sort_values(by = ["gross"], inplace = False, ascending = False)
```

```python
# REGRESSION (BUDGET VS GROSS)
sns.regplot(x = "budget", y  = "gross", data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
# CORRELATION
corr_matrix = df[["year","score","votes","budget","gross","runtime"]].corr(method = "pearson")
sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix for Numerical Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# CONVERTING DATASET TO NUMERICAL
df_numerical = df.copy()
for i in df_numerical.columns:
    if (df_numerical[i].dtype == "object"):
        df_numerical[i] = df_numerical[i].astype("category")
        df_numerical[i] = df_numerical[i].cat.codes

df_numerical
```

```python
# CORRELATION FOR DATASET
corr_matrix_numerical = df_numerical.corr(method = "pearson")
sns.heatmap(corr_matrix_numerical, annot = True)
plt.title("Correlation Matrix for Numerical Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# CORRELATION SORTED
corr_numerical_pairs = df_numerical.corr().unstack().sort_values()
corr_numerical_pairs[(corr_numerical_pairs > 0.5)]
```
